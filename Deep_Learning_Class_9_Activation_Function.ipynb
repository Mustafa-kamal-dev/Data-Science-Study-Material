Activation functions:

- nonlinearity
- ability to learn complex patterns
- decision boundaries
'brain-like' behavior

1. ReLU (Rectified linear unit)
2. Sigmoid
3. Tanh (Hyperbolic tangent)
4. Softmax
5. Leaky ReLU .... (ELU, GELU,.....)

1. ReLU:

if input is +ve, keep it

If input is negative, make it zero

Mostly used in hidden layers

Prole: some neurons get stuck at 0 forever

2. Sigmoid (Output is always between 0 and 1)

anything you put in - becomes a probability

used for binary classification output layer

not recommended for hidden layers

3. Tanh (Output -1 and +1)

4. Softmax

Takes a list of numbers -> converts into a list of probabilities

Total is always = 1.0

[2.1, 1.3, 0.4] = sum = 3.8

[0.552, 0.342, 0.105] = sum = 0.998

Used in output layer of multiclass classification

NLP classification

image classification(CIFAR-10)

5. Leaky ReLU (and friends)

relu problem: if neuron gets negative values forever, it dies

leaky relu fixes this problem

if x<0 -> it gives a small negative slope (0.01x)



```python
from keras.layers import LeakyReLU

model = Sequential([
    Dense(64), LeakyReLU(alpha=0.01)
])
```
