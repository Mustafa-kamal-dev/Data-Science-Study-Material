```python
from sklearn.datasets import make_circles
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
X,y = make_circles(n_samples=200, random_state=0, noise=0.1)
plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm')
plt.title('Non-linear Data - Circular Pattern')
plt.show()
```

```python
model = SVC(kernel='rbf', C=1)
model.fit(X,y)
#rbf (Radial Basis Function) automatically maps data into a higher dimension where it becomes linearly separable
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(X)
print("RBF Kernel Accuracy", accuracy_score(y, y_pred))
```

```python
plt.scatter(X[:,0], X[:,1], c=y_pred, cmap='bwr')
plt.title('SVM with RBF Kernel')
plt.show()
```

```python
from sklearn.datasets import make_moons
X,y = make_moons(n_samples=200, noise=0.1, random_state=0)
model = SVC(kernel='rbf', C=1, gamma=0.5)
model.fit(X,y)
y_pred = model.predict(X)
print('Accuracy:', accuracy_score(y, y_pred))
```

```python
plt.scatter(X[:,0], X[:,1], c=y_pred, cmap='coolwarm')
plt.title('SVM with RBF - Moon')
plt.show()
```

```python
X,y = make_circles(n_samples=200, factor=0.5, noise=0.05)
model =SVC(kernel='poly', degree=3, C=1)
model.fit(X,y)
y_pred=model.predict(X)
print('Accuracy:',accuracy_score(y, y_pred))
plt.scatter(X[:,0],X[:,1], c=y_pred, cmap='coolwarm')
plt.title('SVM with Polynomial Kernel')
plt.show()
```
