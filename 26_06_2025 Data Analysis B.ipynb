```python
df['Revised_Salary'] = df['Salary'] * 1.10

df['Additional cost'] = df['Revised_Salary'] -df['Salary']

department_impact = df.groupby('Department')['Additional cost'].sum()

department_biggest_impact = department_impact.sort_values(ascending=False)


print("Total_additional_cost by Department")

print(department_biggest_impact)

print("Department with the biggest impact")

print(department_biggest_impact.idxmax() , '-->' , department_biggest_impact.max())

```
