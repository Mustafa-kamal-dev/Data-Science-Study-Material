```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sharmaroshan/Heart-UCI-Dataset/master/heart.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.columns= ['Age', 'Sex', 'ChestPainType', 'RestBP', 'Cholesterol', 'FBS' , 'RestEC',
             'MaxHR', 'ExerciseAngina', 'Oldpeak', 'ST_Slope', 'NumVesselsFluoro', 'Thal', 'Target']
```

```python
df.tail(3)
```

```python
sns.countplot(data=df, x='Target')
plt.title('Heart Disease')
plt.xticks([0, 1], ['No Disease', 'Disease'])
plt.show()
```

```python
plt.rcParams['figure.figsize']=(5,3)
```

```python
sns.countplot(data=df, x='Target')
plt.title('Heart Disease')
plt.xticks([0, 1], ['No Disease', 'Disease'])
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Target', multiple='dodge')
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(data=df, x='ChestPainType', hue='Target')
plt.title('Chest Pain Type')
plt.show()
```

```python
sns.histplot(data=df, x='MaxHR', hue='Target', multiple='dodge')
plt.title('Max HR vs Heart Disease')
plt.show()
```

```python
df.groupby('Target')['MaxHR'].mean()
```

```python
df.groupby('Target')['Cholesterol'].mean()
```

```python
sns.histplot(data=df, x='Cholesterol', hue='Target', multiple='dodge')
plt.title('Cholesterol vs Heart Disease')
plt.show()
```

```python
df[['Cholesterol','Target']].groupby('Target').describe()
```

```python
sns.histplot(data=df, x='RestBP', hue='Target', multiple='dodge')
plt.show()
```

```python
sns.countplot(data=df, x='Sex', hue='Target')
plt.xticks([0,1],['Female','Male'])
plt.show()
```
