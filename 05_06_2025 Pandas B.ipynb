```python
def bonus(sal):
    if sal >= 200000:
        return sal * 0.5
    elif sal >= 150000:
        return sal * 0.3
    elif sal >= 100000:
        return sal * 0.25
    
    else:
        return sal * 0.1
```

```python
emp['Bonus']  = emp.Salary.apply(bonus)
emp
```

```python
emp['Bonus2'] = emp.Salary.apply(lambda sal:sal*0.5 if sal>=200000
                                 else sal*.3 if sal>=150000
                                 else sal*.1)
emp

```
