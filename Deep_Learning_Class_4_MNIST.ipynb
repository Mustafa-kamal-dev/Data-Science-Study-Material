```python
from keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train[0].shape
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open('two.jpg')
plt.imshow(img)
plt.axis('off')
```

```python
gray = img.convert('L')
plt.imshow(gray, cmap='gray')
plt.axis('off')
```

```python
resized = gray.resize((28,28))
plt.imshow(resized, cmap='gray')
plt.axis('off')
```

```python
import numpy as np
arr = np.array(resized)
arr_norm = arr / 255.0
plt.imshow(arr_norm, cmap='gray')
plt.axis('off')
```

```python
flat = arr_norm.flatten().reshape(1, -1)
flat.shape
```

```python
from keras import models, layers

model = models.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
  ])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(X_train/255.0, y_train, epochs=6)
```

```python
img = Image.open('nine.JPG').convert('L').resize((28,28))
flat = np.array(img) / 255.0
flat = flat.flatten().reshape(1,28,28)
pred = model.predict(flat)
print('Predicted Digit:', pred.argmax())
```
