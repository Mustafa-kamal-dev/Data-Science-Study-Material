Q: Can we predict whether someone has heart disease based on the health measurements?

BINARY CLASSIFICATION Predict one of the two possible categories 0 or 1, Yes or No, True or False

MULTI CLASS CLASSIFICATION Predict one of more than two categories
Example: What type of animal is this? (Cat, Dog, Horse, Donkey)

REGRESSION: Predict a continuous numerical value
Example: Predict the selling price of a house

MULTI LABEL CLASSIFICATION
Predict multiple binary labels for each instance
Example: Predict which genres apply to a movie (Action, SciFi, Thriller, Drama)

CLUSTERING: Group similar items together without labeled targets (unsupervised learning)
Example: Group customers by spending behaviour

ANOMALY DETECTION: Find unusual data points that do not fit normal patterns.
Example: Detecting credit card fraud

TIME SERIES FORECASTING Predict values over time
Example: Forecast stock prices / electricity usage over days,weeks,months

Heart Dataset:
Type: Supervised learning
Problem: Binary Cllassification Problem
Goal: Predict Target (1: Disease, o:No Disease)
Input: X: 8 columns (features) (Age,Thal,MaxHR,Chol.....)
Output: y: Target

ML Pipeline

1. Data (dataset, scraping, API, Google sheet, JSON, SQL)
2. EDA (shape, missing values, outliers, distribution)
3. Cleaning (fix data types, encode categories)
4. Features (create new features)
5. Split (Features (x) and Outcome(y))
6. Model (Use ML algorithm to learn patterns)
7. Evaluate (Measure model accuracy, precision, recall, confusion matrix)
