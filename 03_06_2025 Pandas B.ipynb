```python
emp = pd.read_excel('employee.xlsx')
emp
```

```python
emp['Increment'] = [50000 if age >= 60
                    else 40000 if age >= 50
                    else 30000 if age >= 40
                    else 20000 if age >= 30
                    else 10000 if age >= 20
                    else 5000 for age in emp.Age]
emp
```

```python
emp['Bonus'] = [sal*0.50 if sal>=200000
                else sal*0.25 if sal>=150000
                else sal*0.15 if sal>=75000
                else sal*0.10
                for sal in emp.Salary]
emp
```
